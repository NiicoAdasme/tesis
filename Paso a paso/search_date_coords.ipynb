### Fecha y coordenadas a buscar

### Debido a que las posicones de la variable time son indices que corresponden a las posiciones considerando la fecha de inicio como la primera posicion hasta el a√±o 2019 segun el archivo netcdf ingresado, utilizamos un rango para encontrar la posicion que coincida con la fecha indicada

```python
date_range = pd.date_range(start= starting_date, end= ending_date, freq= 'M')
pos_date = date_range.size
print(date_range)
print('Posicion a buscar: ', pos_date)
```
